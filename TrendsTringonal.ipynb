### ENERGY ERROR

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mother_dir = './TRIGONAL/'

title = f'$\Delta$H w.r.t. USPEX GS with INITIAL SET'
fig, ax = plt.subplots(1, 1, figsize=(15, 7))
fig.subplots_adjust(hspace=0.0)
fig.suptitle(title, fontsize=20, y=0.95, fontweight='bold')
max_x_ticks = np.array([])

complist = [1,2,3,4]
for comp in complist:
    dir_name = f'{comp}'
    df_tot = pd.read_csv(mother_dir + f'{dir_name}/TotalStaticsBefore.csv', header=None, na_filter=False, index_col = 0)
    tot = (df_tot.to_numpy()).T

    x = np.arange(1, len(tot[0])+1) 

    ax.plot(x, tot[0] , marker = 'o', markeredgecolor='white', markersize=6, label=f'A$_{comp}$B')
    max_x_ticks = np.append(max_x_ticks, len(tot[0]))


ax.set_ylabel(r'$\Delta$H [eV/atom]', fontsize=22)
ax.yaxis.set_label_coords(-0.07, 0.5)

step = 0.05
y_ticks  = np.arange(0.0,0.5,step, dtype=float)
y_ticks = np.round(y_ticks, decimals=2)
ax.set_yticks(y_ticks)
ax.set_yticklabels([f'{i:.2f}' for i in y_ticks], fontsize=18)
ax.set_ylim(0,0.4)

ax.set_xlabel('Size initial set', fontsize=22)
ax.xaxis.set_label_coords(0.5, -0.10)

xticks = np.arange(1, max(max_x_ticks)+1, 5)
xticks = np.round(xticks, decimals=0) 
ax.set_xticks(xticks)
ax.set_xlim(0,max(max_x_ticks)+1)
ax.set_xticklabels([f'{i:.0f}' for i in xticks], fontsize=15, rotation=45)

ax.grid(True, ls=':')
ax.legend(loc='upper right', fontsize=15)

plt.show()
fig.savefig('FIG/Trigonal_INIT_H.png', bbox_inches="tight")
```

### SIZE FINAL SET

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mother_dir = './HvsINIT/'

complist = [1, 2, 3, 4]

title = f'SIZE FINAL SET with threshold 0.9'
fig, ax = plt.subplots(1, 1, figsize=(15, 7))
fig.subplots_adjust(hspace=0.0)
fig.suptitle(title, fontsize=20, y=0.95, fontweight='bold')

max_x_ticks = np.array([])

for comp in complist:
    dir_name = f'{comp}'
    df_numtemp = pd.read_csv(mother_dir + f'{dir_name}/NumberTempRedu.csv', header=None, na_filter = False, index_col = 0)
    numtemp = (df_numtemp.to_numpy()).T
    x = np.arange(1, len(numtemp[0])+1)

    ax.plot(x, numtemp[0], marker = 'o', markeredgecolor='white', markersize=6, label=f'A$_{comp}$B')
    max_x_ticks = np.append(max_x_ticks, len(numtemp[0]))

ax.set_ylabel('Size final set', fontsize=22)
ax.yaxis.set_label_coords(-0.07, 0.5)

step = 2
y_ticks  = np.arange(0,30,step, dtype=float)
y_ticks = np.round(y_ticks, decimals=0)
ax.set_yticks(y_ticks)
ax.set_yticklabels([f'{i:.0f}' for i in y_ticks], fontsize=18)
ax.set_ylim(0,30)

ax.set_xlabel('Size initial set', fontsize=22)
ax.xaxis.set_label_coords(0.5, -0.10)

xticks = np.arange(1, max(max_x_ticks)+1, 5)
xticks = np.round(xticks, decimals=0) 
ax.set_xticks(xticks)
ax.set_xlim(0,max(max_x_ticks)+1)
ax.set_xticklabels([f'{i:.0f}' for i in xticks], fontsize=15, rotation=45)

ax.grid(True, ls=':')
ax.legend(loc='lower right', fontsize=15)

plt.show()
fig.savefig('FIG/Trigonal_INIT_Ntemp.png', bbox_inches="tight")
```

### COMPARISON TRIGONAL AND TETRAGONAL

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np






complist = [1,2,3,4]
for comp in complist:
    fig, ax = plt.subplots(1, 1, figsize=(15, 7))
    fig.subplots_adjust(hspace=0.0)
    title = r'$\Delta$H w.r.t. USPEX GS with INITIAL SET' + f' A{comp}B'  
    fig.suptitle(title, fontsize=20, y=0.95, fontweight='bold')
    max_x_ticks = np.array([])

    dir_name = f'{comp}'
    mother_dir = './HvsINIT/'
    df_tot = pd.read_csv(mother_dir + f'{dir_name}/TotalStaticsBefore.csv', header=None, na_filter=False, index_col = 0)
    tot = (df_tot.to_numpy()).T

    x = np.arange(1, len(tot[0])+1) 

    ax.plot(x, tot[0] , marker = 'o', markeredgecolor='white', markersize=9, label=f'Tetragonal', color='green')
    max_x_ticks = np.append(max_x_ticks, len(tot[0]))

    mother_dir = './TRIGONAL/'
    df_tot = pd.read_csv(mother_dir + f'{dir_name}/TotalStaticsBefore.csv', header=None, na_filter=False, index_col = 0)
    tot = (df_tot.to_numpy()).T

    x = np.arange(1, len(tot[0])+1) 

    ax.plot(x, tot[0] , marker = '*', markeredgecolor='white', markersize=15, label=f'Trigonal', color='red', linestyle='--')
    max_x_ticks = np.append(max_x_ticks, len(tot[0]))

    mother_dir = './CUBIC/'
    df_tot = pd.read_csv(mother_dir + f'{dir_name}/TotalStaticsBefore.csv', header=None, na_filter=False, index_col = 0)
    tot_cub = (df_tot.to_numpy()).T

    x = np.arange(1, len(tot_cub[0])+1) 

    ax.plot(x, tot_cub[0] , marker = 'h', markeredgecolor='white', markersize=10, label=f'Cubic', color='blue', linestyle=':')
    max_x_ticks = np.append(max_x_ticks, len(tot_cub[0]))


    ax.set_ylabel(r'$\Delta$H [meV/atom]', fontsize=22, family='serif')
    ax.yaxis.set_label_coords(-0.07, 0.5)

    step = 0.1
    y_ticks  = np.arange(0.0,tot_cub[0].min() + 0.5,step, dtype=float)
    y_ticks = np.round(y_ticks, decimals=2)
    ax.set_yticks(y_ticks)
    ax.set_yticklabels([f'{i*1000:.0f}' for i in y_ticks], fontsize=18)
    ax.set_ylim(0,tot_cub[0].min() + 0.5 )

    ax.set_xlabel('Size initial set', fontsize=22, family='serif')
    ax.xaxis.set_label_coords(0.5, -0.10)
    step_x = 5
    if comp == 2:
        step_x = 10
    xticks = np.arange(0, max(max_x_ticks), step_x)
    xticks = np.round(xticks, decimals=0) 
    ax.set_xticks(xticks)
    ax.set_xlim(0,max(max_x_ticks)+1)
    ax.set_xticklabels([f'{i:.0f}' for i in xticks], fontsize=15)

    ax.grid(True, ls=':')
    ax.legend(loc='upper right', fontsize=15)

    plt.show()
    fig.savefig(f'FIG/COMPARISON_SG_A{comp}B.png', bbox_inches="tight")
```
