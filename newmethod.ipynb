```python
import random

clusters = {
    
    
    '1' : {
        'couples' : ['ClB', 'NB', 'CaLi'],
        'freq' : 0.05,
    },
    '2' : {
        'couples' : ['BeNa', 'OLi', 'MgNa', 'AlLi', 'SiLi', 'PLi', 'SLi', 'ClLi', 'ArLi'],
        'freq' : 0.05,
    },
    '3' : {
        'couples' : [],
        'freq' : 0.9,
    },
    #need to add the cluster of couples out of clusters
}
clusters = {}
import json
with open('clusters.json', 'r') as f:
    clusters = json.load(f)

banned = []
couples = []
couples_in_clusters = [] 

original_clusters = []
freq_cluster = []

for cluster in clusters.keys():
    couples_in_clusters.append([])
    original_clusters.append((int(cluster)-1,clusters[cluster]['couples']))
    freq_cluster.append(clusters[cluster]['freq'])

while len(couples) < 105:
    idx_empty_clusters = []
    for idx, cluster in enumerate(couples_in_clusters):
        if len(cluster) == 0:
            idx_empty_clusters.append(idx)


    if len(idx_empty_clusters) != 0:
        freq_empty_clusters = [freq_cluster[x] for x in idx_empty_clusters]
        rnum = random.uniform(0, sum(freq_empty_clusters))
        for idx, freq in enumerate(freq_empty_clusters):
            if rnum < freq:
                possible_choice = [x for x in original_clusters[idx_empty_clusters[idx]][1] if x not in couples]

                if len(possible_choice) == 0:
                    del couples_in_clusters[idx_empty_clusters[idx]]
                    del freq_cluster[idx_empty_clusters[idx]]
                    del original_clusters[idx_empty_clusters[idx]]
                    break
                new_couple = random.choice(possible_choice)
                couples.append(new_couple) # to add in update
                couples_in_clusters[idx_empty_clusters[idx]].append(new_couple) # to add in update
                break
            else:
                rnum -= freq
    else:
        rnum = random.uniform(0, sum(freq_cluster))
        for idx, freq in enumerate(freq_cluster):
            if rnum < freq:
                possible_choice = [x for x in original_clusters[idx][1] if x not in couples]

                if len(possible_choice) == 0:
                    del original_clusters[idx]
                    del freq_cluster[idx]
                    break
                new_couple = random.choice(possible_choice)
                couples.append(new_couple) # to add in update
                couples_in_clusters[original_clusters[idx][0]].append(new_couple) # to add in update
                break
            else:
                rnum -= freq
        


```
